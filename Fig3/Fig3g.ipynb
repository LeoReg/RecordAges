```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from numba import jit
import pandas as pd

def modlin(x,a,b):
    return a*x+b
def sep(l):
    s1,s2='',''
    u=0
    for k in range(len(l)):
        if l[k]!=' ' and u==0:
            s1+=l[k]
        if l[k]!=' ' and u==1:
            s2+=l[k]
        if l[k]==' ':
            u=1
    return s1,s2


P=pd.read_csv("london_temperature.txt")
for k in P:
  Val=P[k]




#Import des donnes et changement des dates au format timedelta64
Dates,Values=[],[]
for l in Val:
    s1,s2=sep(l)
    Dates.append(s1)
    Values.append(np.float64(s2))


Dates=np.array([np.int64(np.datetime64(d)-np.datetime64(Dates[0])) for d in Dates])
Values=Values[:]

Indices=[0]

for k in range(100,len(Values)-100):
    if Values[k]==max(Values[k-100:k+100]):
        Indices.append(k)
    if Values[k]==min(Values[k-100:k+100]):
        Indices.append(k)
Indices+=[len(Values)-1]
b=Values[0]


for k in range(len(Indices)-1):
     def modlin(x,a):
        return a*x+b
     x,y=Dates[Indices[k]:Indices[k+1]]-Dates[Indices[k]],Values[Indices[k]:Indices[k+1]]
     print(y)
     sol=curve_fit(modlin,x,y)
     a=sol[0]
     Values[Indices[k]:]-=modlin(Dates[Indices[k]:]-Dates[Indices[k]],a)
     b=Values[Indices[k+1]-1] 


plt.plot(Values)
plt.show()


plt.plot(np.cumsum(Values-np.mean(Values)))
plt.show()

Values=np.cumsum(Values-np.mean(Values))

```

```python
H=0.8
dw=1/H
theta=1-H

from numba import jit
@jit(nopython=True)
def tau(V):
    T=[]
    M=V[0]
    tau=1
    for k in range(1,len(V)-1):
        if V[k]>M:
            T.append(tau)
            tau=1
        else:
            tau+=1
        M=max(V[k],M)
    return T


T=[]
epochs=1000
window=len(Values)//epochs
print(window)
for k in range(0,epochs):
  T.append(tau(Values[k*window:]))#(k+2)*window]))

Tn=T

@jit(nopython=True)
def FreCum(T,Tl):
    S=[]
    for t in Tl:
        S.append(np.sum(T>=t)/len(T))
    return S

##
Tl=np.unique(np.int64(np.logspace(0.3,2,15)))
Tl=np.float64(Tl)

dw=1/H
i=0



N=[1,2,3]
Marker=['o','*','s']
i=-1
for n in N:
    i+=1
    LT=[]
    for t in Tn:
      if len(t)>n:
        LT.append(t)
    Times=np.logspace(0.3,2,13)
    Ti=[LT[k][n] for k in range(len(LT))]
    Ti=np.array(Ti)
    Ti=Ti[Ti>0]
    print(len(Ti))
    Sl=np.array([np.sum(Ti>=ti)/len(Ti) for ti in Times])
    n0=n
    plt.plot(Times/n0**dw,Sl*n0,Marker[i],markersize=10,linestyle='',fillstyle='none',linewidth=20,markeredgewidth=2.0)
    if n==N[-1]:
      Ti=np.logspace(0,2,41)
      plt.plot(Ti/n0**dw,(Ti/Times[len(Times)//4])**(-H)*Sl[len(Times)//4]*n0,'--r',linewidth=4)
    if n==N[0]:
      Ti=np.logspace(0,2,41)  
      plt.plot(Ti/n0**dw,(Ti/Times[len(Times)//4*3])**(-theta)*Sl[len(Times)//4*3]*n0,'--k',linewidth=4)

plt.xlabel(r'$\tau/n^{d_w}$',fontsize=20)
plt.ylabel(r'$nS(n,\tau)$',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.semilogx()
plt.semilogy()

plt.savefig('Records_Temp_London.svg')




```
